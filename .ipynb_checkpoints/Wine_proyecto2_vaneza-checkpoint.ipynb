## Identificar y eliminar columnas y registros innecesarios

```python
# Eliminar la columna Unnamed 
# Reemplazar el indentificador secuencial del dataset por la columna Unnamed.
df=pd.read_csv ("wine_reviews_dataset.csv", index_col="Unnamed: 0")
df
```

```python
# Eliminar filas duplicadas
# Identificar filas duplicadas en el df
df_filas_duplicadas = df[df.duplicated()]
print(df_filas_duplicadas)
```

```python
# Visualizar la cantidad de registros duplicados
df.duplicated().sum()
```

```python
#Eliminar registros duplicados
df_sin_dupli=df.drop_duplicates ()
df_sin_dupli
```

```python
# Número de registros del df sin duplicados 
len(df_sin_dupli)
```

```python
# Identificar la cantidad de los valores nulos en el df
df_sin_dupli.isnull().sum()
```

```python
# Identificar el promedio de los valores nulos en el df
df_sin_dupli.isnull().mean()
```

```python
# Se ordena por  % nulos
df_sin_dupli.isnull().mean().sort_values(ascending = False)
```

```python
# El origen de la denominación del vino proviene de las regiones. El análisis del vino debe por lo menos tener una región 
# asignada.
# Comparando la region_1 y region_2 
# Filtrar los registros que presentan NULOS en la region_1
df_r1_sin_dupli=df_sin_dupli[df_sin_dupli.isnull().region_1]
df_r1_sin_dupli
```

```python
# Longitud del df con registros que presentan NULOS en la region_1
len(df_r1_sin_dupli)
```

```python
# Obs: Posiblemente los datos de la región_1 son nulos en la region_2
```

```python
# Filtrar los registros que presentan NULOS en la region_2
df_r2_sin_dupli=df_sin_dupli[df_sin_dupli.isnull().region_2]
df_r2_sin_dupli
```

```python
# Longitud del df con registros que presentan NULOS en la region_2
len(df_r2_sin_dupli)
```

```python
# Hipótesis: Posiblemente todos los datos de la región_2 son nulos cuando la region_1 presenta datos nulos.
```

```python
# Filtrar registros nulos en region_1 y region_2
df_sin_dupli_r1r2=  df_sin_dupli[ df_sin_dupli.isnull().region_1 & df_sin_dupli.isnull().region_2 ]
df_sin_dupli_r1r2
```

```python
# Longitud del df con registros que presentan NULOS en la region_1 y region_2
len(df_sin_dupli_r1r2) 
```

```python
# Filtrar bloques de registros 
Filtrado5=df_sin_dupli_r1r2.loc[100000:129966, 'region_1':'region_2' ]
Filtrado5
```

```python
# Se confirma la hipótesis: región_1 tiene valores nulos en region_2.
# Se recomienda eliminar los 19560 registros de region_1=Nulo que representa el 16.30 % del  df_sin_dupli
```

```python
#Eliminar registros nulos de la region_1
df_sin_dupli_2=df_sin_dupli.dropna(subset=["region_1"])
df_sin_dupli_2
```

```python
len(df_sin_dupli_2)
```
